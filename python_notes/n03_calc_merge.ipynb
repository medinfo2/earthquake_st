```python
import pandas as pd
import pickle
```

```python
import datetime

dt_now = datetime.datetime.now()
ts = dt_now.strftime('%Y%m%d_%H%M%S_%f')
ts
```

```python
def pickle_load(path):
    with open(path, mode='rb') as f:
        data = pickle.load(f)
        return data
```

```python
f_desmat = "../dat/design_matrix_pre_analysis.pkl"
df_desmat = pd.read_pickle(f_desmat)\
    .assign(number = lambda df:df.number.astype(int))

df_desmat.head(50)
```

```python
# f_calc = "../dat/d_res_temp_20231031_105443_337076.pkl"
# f_calc = "../dat/d_res_temp_20231108_104428_897264.pkl"
# f_calc = '../dat/d_res_temp_20231127_145813_118949.pkl'
f_calc = '../dat/d_res_temp_20231229_175540_583251.pkl'
ds = pickle_load(f_calc)
```

```python
%%time
l_part = []
needed_vals = {"Index",'ot_abs_max_acc'}
for k,v in ds.items():
    # print(k)
    # print(v.keys())
    d_part = dict()
    d_part["Index"] = v["Index"]
    d_part["experiment_number"] = v["experiment_number"]
    d_part["safetybar"] = v["備考"]
    d_part["ot_abs_max_acc_val"] = v["ot_abs_max_acc"]["max_abs"]
    d_part["ot_max_dif_vel_val"] = v["ot_max_dif_vel"]["d_max_dif"]["abs_sub_min_max"]
    
    d_part["ot_abs_max_acc_val_whole"] = v["ot_abs_max_acc_whole"]["max_abs"]
    d_part["ot_max_dif_vel_val_whole"] = v["ot_max_dif_vel_whole"]["d_max_dif"]["abs_sub_min_max"]
        
    d_part["m_abs_max_acc_val"] = v["m_abs_max_acc"]["max_abs"]
    d_part["m_max_dif_vel_val"] = v["m_max_dif_vel"]["d_max_dif"]["abs_sub_min_max"]
    
    d_part["m_abs_max_acc_val_whole"] = v["m_abs_max_acc_whole"]["max_abs"]
    d_part["m_max_dif_vel_val_whole"] = v["m_max_dif_vel_whole"]["d_max_dif"]["abs_sub_min_max"]

    
    d_part["sty_abs_max_acc_val"] = v["sty_abs_max_acc"]["max_abs"]
    d_part["sty_max_dif_vel_val"] = v["sty_max_dif_vel"]["d_max_dif"]["abs_sub_min_max"]
    
    d_part["sec_start"] = v["sec_start"]
    d_part["sec_end"] = v["sec_end"]
    # l_part.append({k: v for k,v in v.items() if k in needed_vals}) 
    l_part.append(d_part)
    # break
```

```python
df_res = pd.DataFrame().from_dict(l_part)\
    .rename(columns={"Index":"number"})\
    .join(df_desmat, on="number", how="left", lsuffix="_res", rsuffix="_des")
df_res
```

```python
f_out ="../dat/df_des_res_fix_{}.csv".format(ts)
df_res.to_csv(f_out)
print(f_out)
```
